```python
# Function to extract title and paragraphs of webpage
def extract_text(url):
    # making requests instance
    reqs = requests.get(url)

    # using the BeautifulSoup module
    soup = BeautifulSoup(reqs.text, 'html.parser')

    # displaying the title
    titles = ""
    for title in soup.find_all('title'):
        titles = titles + " " + title.get_text()

    paragraphs = ""
    for paragraph in soup.find_all('p'):
        paragraphs = paragraphs + " " + paragraph.get_text()
    
    #we will join title and text with period so that title can be considered a seperate sentence
    complete_text = titles + "." + paragraphs
    
    return complete_text
    
    
    
```

# Data Cleaning

```python
#function to decontract words
# https://stackoverflow.com/a/47091490/4084039
import re

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
# Function to preprocess and clean text 
from tqdm import tqdm
def preprocess_text(text_data, stop_words):
    preprocessed_text = []
    # tqdm is for printing the status bar
    for sentence in tqdm(text_data):
        if type(sentence) != str:
            sent = ' '
        else:
            sent = decontracted(sentence)
            sent = sent.replace('\\r', ' ')
            sent = sent.replace('\\n', ' ')
            sent = sent.replace('\\"', ' ')
            sent = re.sub('[^A-Za-z0-9.]+', ' ', sent)
        # https://gist.github.com/sebleier/554280
        sent = sent.split()
        sent = [word for word in sent if word not in stop_words]
        sent = ' '.join(sent)
        preprocessed_text.append(sent.lower().strip())
    return preprocessed_text
```

## Number of complex words

```python
# Load the CMU Pronouncing Dictionary
pronouncing_dict = cmudict.dict()
```

```python
# Define a function to count syllables in a word we will consider complexity of a word based on number of syllables in it
def count_syllables(word):
    if word.lower() in pronouncing_dict:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in pronouncing_dict[word.lower()]])
    else:
        return 0

#following function returns count of complex words and percentage of complex words     
def get_complex_words(text, complex_threshold):
    words = nltk.word_tokenize(text)
    complex_word_count = sum(1 for word in words if count_syllables(word) >= complex_threshold)
    return complex_word_count, np.round(complex_word_count/len(words),4)  
```

```python
count_of_complex_words = []
perc_of_complex_words = []
complex_threshold = 2
for text in all_texts_preprocessed:
    count, perc = get_complex_words(text, complex_threshold)
    count_of_complex_words.append(count)
    perc_of_complex_words.append(perc)
```

```python
# Define a list of personal pronoun tags
personal_pronoun_tags = ['PRP', 'PRP$', 'WP', 'WP$']
```

```python
# Create a function to count personal pronouns
def count_personal_pronouns(text):
    pronoun_count = 0
    pronoun_list = []
    for sentence in text:
        pos_tags = pos_tag(sentence)
        pronoun_count += sum(1 for word, pos in pos_tags if pos in personal_pronoun_tags)
        pronoun_list.extend([word for word, pos in pos_tags if pos in personal_pronoun_tags])
        
        
    return pronoun_count, list(set(pronoun_list))
```

```python
count_personal_pronouns(words)
```

```python
pronoun_count = []
pronoun_list = []
for text in all_texts_preprocessed:
    sentences = sent_tokenize(text)
    words = [word_tokenize(sentence) for sentence in sentences]
    a,b = count_personal_pronouns(words)
    pronoun_count.append(a)
    pronoun_list.append(b)
```

```python
fig,ax = plt.subplots(1,1, figsize=(10,7))
sns.boxplot(df["AVG SYLLABLE PER WORD"], ax = ax)
ax.set_ylabel('density', fontdict = {'fontsize': 15})
ax.set_title('AVG SYLLABLE PER WORD', fontdict = {'fontsize': 20})
plt.legend()
plt.show()
```
